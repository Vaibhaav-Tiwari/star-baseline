# StaR — Dataset Exploration
Load synthetic DLPFC dataset, visualise spatial layout, expression heatmap.

```python
import sys; sys.path.insert(0, '..')
import numpy as np
import matplotlib.pyplot as plt
from star.data import make_synthetic_dlpfc
from star.visualize import plot_spatial_domains

ds = make_synthetic_dlpfc(n_spots=3000, n_genes=3000, n_layers=7, seed=0)
print(f'Spots: {ds.n_spots} | Genes: {ds.n_genes} | Layers: {ds.n_layers}')
print(f'Edge count: {ds.edge_index.shape[1]}')
```

```python
plot_spatial_domains(ds, 'figures/spatial_domains.png')
from IPython.display import Image; Image('figures/spatial_domains.png')
```

```python
# Per-layer expression statistics
import pandas as pd
rows = []
for layer in ds.unique_layers:
    mask = ds.labels == layer
    rows.append({'Layer': layer, 'N spots': mask.sum(),
                 'Mean expr': ds.X[mask].mean().round(4),
                 'Std expr':  ds.X[mask].std().round(4)})
pd.DataFrame(rows)
```
